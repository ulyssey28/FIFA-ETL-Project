```python
import pandas as pd
from sqlalchemy import create_engine, inspect

import pymysql
pymysql.install_as_MySQLdb()
```

```python
csv_file = "Resources/data.csv"
fifa_df = pd.read_csv(csv_file)
fifa_df.head()
```

```python
player_df = fifa_df.rename(columns={"Nationality":"country"})

re_player_df = player_df.rename(columns={"Loaned From":"Loaned"})
re_player_df['Name'] = re_player_df['Name'].astype(str)

re_player_df.head()
```

```python
json_file = "Resources/fifa_ranking.json"
rank_df = pd.read_json(json_file)
rank_df.head()
```

```python
rank_2018_df = rank_df.loc[rank_df['rank_date'] >= "2018-06-01"]
rank_2018_df
```

```python
new_rank_2018_df = rank_2018_df[["country_full", "rank", "rank_change"]]
new_rank_2018_df = new_rank_2018_df.rename(columns={"country_full":"country"});
new_rank_2018_df.head()
```

```python
re_player_df['Wage'].head()
wage_list = []
for wage in re_player_df['Wage']:
    wages = wage.translate({ord(i): None for i in "€K"})
    wage_list.append(wages)

re_player_df['Wage'] = pd.Series(wage_list)
re_player_df['Wage'] = re_player_df['Wage'].astype("int")

re_player_df['Wage'] = re_player_df['Wage'] * 1000
```

```python
re_player_df['Wage'].head()
```

```python
Top_player_df = re_player_df.loc[re_player_df['Overall'] >= 75]
Top_player_df.head()
```

```python
rds_connection_string = "root:{PASSWORD}@127.0.0.1/fifa_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
re_player_df.to_sql(name='player_data', con=engine, if_exists='append', index=False)
```

```python
new_rank_2018_df.to_sql(name='country_rankings', con=engine, if_exists='append', index=False)
```
